```python
import matplotlib.pyplot as plt
import numpy as np
import shapely.geometry
from shapely.geometry import Polygon
```

```python
# read flash.dat to a list of lists
sample_data= [(i.strip().split()) for i in open("amplitude.dat").readlines()]
sample_data=np.array(sample_data, dtype=float)
```

```python
data=np.array([0.0 for i in range(len(sample_data))],dtype=float)
for s in range(len(sample_data)):
    data[s]=(sample_data[s][0])
```

```python
print(sample_data)
print(data)
```

```python
x=data[:-8]
y=data[8:]
fig,ax=plt.subplots(figsize=(10,8))
ax.plot(x,y)
for t in range(-20,20):
    ax.plot(np.full(40,t), np.arange(-20,20,1), alpha=0.3)
    ax.plot(np.arange(-20,20,1), np.full(40,t), alpha=0.3)
fig.show()
```

```python
# (xmin, y where) (xmax, y where) (x where, y min) (x where y max)
y_ = [np.min(y), np.max(y)]
x_=[np.min(x), np.max(x)]
print(((x_[1]-x_[0])**2)/2 * (0.5*0.5))
print(y_)
coords=((x_[0], y[np.where(x[x==x_[0]])[0]]),
        (x_[1], y[np.where(x[x==x_[1]])[0]]),
        (y_[0], x[np.where(y[y==y_[0]])[0]]),
        (y_[1], x[np.where(y[y==y_[1]])[0]]))

p=Polygon(coords)

print(f"{p.area}")
```
