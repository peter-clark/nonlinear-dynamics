```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic_map(x0, n, r):
    xn= np.array([0.0 for x in range(n)], dtype=float)
    xn[0]=x0
    for i in range(n-1):
        xn[i+1]=xn[i]*r*(1-xn[i])
    return xn
```

```python
def bifurcation_slice(plot, x0, rmin, rmax, rstep, n, k):
    r = 0
    r_range=(rmax-rmin)/rstep
    for i in range(int(r_range)): # how many slices
        r=rmin+(i*rstep)
        current_step=logistic_map(x0, n, r) # get slice
        current_step=current_step[5:] # remove first k items
        plot.scatter(np.repeat(r, n-k), current_step, marker='.', alpha=0.01, color='black') # current r value v slice
```

```python
fig, ax=plt.subplots(figsize=(10,8))

x0=0.2
rmin=2.4
rmax=4 
rstep=0.01 
n=1000
k=5

bifurcation_slice(ax, x0, rmin, rmax, rstep, n, k)

plt.show()
```
